```python
#menambahkan library
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
#untuk membaca gambar
img = imageio.imread("domba.jpg")
```

```python
img_height = img.shape[0] #untuk mengatur ukuran vertikal
img_width = img.shape[1] #untuk mengukur ukuran horizontal
img_channel = img.shape[2] #memanggil channel
img_type = img.dtype
```

```python
img_brightness = np.zeros(img.shape, dtype=np.uint8) #mengindikasikan bahwa array yang dibuat akan digunakan untuk menyimpan informasi tentang kecerahan (brightness) gambar
```

```python
def brighter(nilai): #fungsi ini bertujuan untuk meningkatkan kecerahan (brightness) gambar atau nilai yang diberikan
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 #rumus untuk mengubah rgb ke grayscale
            gray += nilai
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            img_brightness[y][x] = (gray, gray, gray)
```

```python
brighter(-100) #mengatur nilai kecerahan lebih gelap
plt.imshow(img_brightness)
plt.title("Brightness -100")
plt.show()

brighter(100) #mengatur nilai kecerahan lebih terang
plt.imshow(img_brightness)
plt.title("Brightness 100")
plt.show()
```

```python
img_rgbbrightness = np.zeros(img.shape, dtype=np.uint8) #menunjukkan bahwa array yang dibuat akan digunakan untuk menyimpan informasi tentang kecerahan (brightness) dalam ruang warna RGB
```

```python
def rgbbrighter(nilai): #mengindikasikan bahwa fungsi ini berkaitan dengan peningkatan kecerahan (brightness) dalam ruang warna RGB
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            red += nilai
            if red > 255:
                red = 255
            if red < 0:
                red = 0
            green = img[y][x][1]
            green += nilai
            if green > 255:
                green = 255
            if green < 0:
                green = 0
            blue = img[y][x][2]
            blue += nilai
            if blue > 255:
                blue = 255
            if blue < 0:
                blue = 0
            img_rgbbrightness[y][x] = (red, green, blue)
```

```python
rgbbrighter(-100) #mengatur nilai kecerahan lebih gelap
plt.imshow(img_rgbbrightness)
plt.title("Brightness -100")
plt.show()

rgbbrighter(100) #mengatur nilai kecerahan lebih terang
plt.imshow(img_rgbbrightness)
plt.title("Brightness 100")
plt.show()
```

```python
img_contrass = np.zeros(img.shape, dtype=np.uint8) #digunakan untuk menyimpan informasi tentang kontras (contrast) gambar
```

```python
def contrass(nilai): 
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray *= nilai
            if gray > 255:
                gray = 255
            img_contrass[y][x] = (gray, gray, gray)
```

```python
contrass(2)
plt.imshow(img_contrass)
plt.title("Contrass 2")
plt.show()

contrass(3)
plt.imshow(img_contrass)
plt.title("Contrass 3")
plt.show()
```

```python
img_autocontrass = np.zeros(img.shape, dtype=np.uint8) #yang menentukan kontrasnya komputer
```

```python
def autocontrass():
    xmax = 300
    xmin = 0
    d = 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            if gray < xmax:
                xmax = gray
            if gray > xmin:
                xmin = gray
    d = xmin-xmax
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 #rumus
            gray = int(float(255/d) * (gray-xmax))
            img_autocontrass[y][x] = (gray, gray, gray)
```

```python
autocontrass() # mengambil gambar sebagai input dan mengaplikasikan penyesuaian kontras otomatis pada gambar tersebut
plt.imshow(img_autocontrass) # untuk menampilkan gambar
plt.title("Contrass Autolevel") #memberikan judul pada plot atau gambar yang ditampilkan
plt.show() # menampilkan semua plot atau gambar yang sudah diatur sebelumnya

```
